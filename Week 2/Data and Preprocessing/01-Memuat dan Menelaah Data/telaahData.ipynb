### Mencari pencilan dengan Tukey's fences (3)
---

Error ini terjadi karena variabel outlier_filter belum didefinisikan sebelum digunakan. Anda perlu membuat filter outlier untuk kolom 'assists' terlebih dahulu.

```python
# df_noid[outlier_filter['assists']] \
#     .loc[:, ['player_name', 'assists']] \
#     .sort_values(by=['assists'], ascending=False)

def detect_outliers_iqr(data, column):
    """Deteksi outlier menggunakan metode IQR"""
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]

    return outliers, {'Q1': Q1, 'Q3': Q3, 'IQR': IQR, 'lower': lower_bound, 'upper': upper_bound}

# Gunakan fungsi
if 'assists' in df_noid.columns:
    outliers, stats = detect_outliers_iqr(df_noid, 'assists')

    print("=== OUTLIER DETECTION RESULTS ===")
    print(f"Statistik:")
    print(f"  Q1: {stats['Q1']:.2f}")
    print(f"  Q3: {stats['Q3']:.2f}")
    print(f"  IQR: {stats['IQR']:.2f}")
    print(f"  Batas bawah: {stats['lower']:.2f}")
    print(f"  Batas atas: {stats['upper']:.2f}")
    print(f"\nJumlah outlier: {len(outliers)}")

    if not outliers.empty:
        print("\nData outlier:")
        print(outliers[['player_name', 'assists']].sort_values('assists', ascending=False))
```

### Korelasi Pearson antara kolom-kolom numerik
---
* Method corr() menghasilkan tabel korelasi pearson antar kolom-kolom numerik.
* Rentang nilai: antara -1 dan 1
* -1 = korelasi negatif | 0 = tidak ada korelasi linier | +1 = korelasi positif

Error ini terjadi karena mencoba menghitung korelasi (corr()) pada kolom yang mengandung data non-numerik (string). Dari pesan error, terlihat ada nilai 'F' yang tidak bisa dikonversi ke float.

**Penyebab:**

df_noid.loc[:,'games':] mengambil semua kolom dari 'games' sampai kolom terakhir, dan beberapa kolom di antaranya berisi data string/text.

```python
# df_noid.loc[:,'games':].corr()

# Ambil kolom dari 'games' sampai akhir
start_col = 'games'
if start_col in df_noid.columns:
    # Dapatkan posisi index kolom 'games'
    start_idx = df_noid.columns.get_loc(start_col)

    # Ambil semua kolom dari 'games' sampai akhir
    df_subset = df_noid.iloc[:, start_idx:]

    # Pilih hanya kolom numerik
    df_numeric = df_subset.select_dtypes(include=['number'])

    print(f"Kolom dari '{start_col}' sampai akhir: {df_subset.columns.tolist()}")
    print(f"\nKolom numerik: {df_numeric.columns.tolist()}")

    if len(df_numeric.columns) > 1:
        korelasi = df_numeric.corr()
        print("\nMatriks Korelasi:")
        print(korelasi)
    else:
        print("Tidak cukup kolom numerik untuk korelasi")
else:
    print(f"Kolom '{start_col}' tidak ditemukan")
```

**NOTE:**
*   Selalu cek tipe data kolom sebelum menghitung korelasi
*   Korelasi hanya bisa dihitung untuk data numerik
*   Gunakan select_dtypes(include=['number']) untuk memilih kolom numerik saja
