```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = ColumnTransformer([('one_hot_encoder', OneHotEncoder(),[1])], remainder = 'passthrough')
X = onehotencoder.fit_transform(X)
X = X[:, 1:]
```

```python
print(X)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Now let's make the ANN!

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 20)
```

Making predictions and evaluating the model

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred_ = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Menghitung confusion matrix
cm = confusion_matrix(y_test, y_pred_)
print("Confusion Matrix:\n", cm)

# Menghitung metrik performa
accuracy = accuracy_score(y_test, y_pred_)
precision = precision_score(y_test, y_pred_)
recall = recall_score(y_test, y_pred_)
f1 = f1_score(y_test, y_pred_)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

**## Eksperimen Baru: ANN dengan Dropout Layer**

```python
from keras.layers import Dropout

# Initialising the ANN with Dropout
classifier_dropout = Sequential()

# Adding the input layer and the first hidden layer with Dropout
classifier_dropout.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
classifier_dropout.add(Dropout(rate = 0.1))

# Adding the second hidden layer with Dropout
classifier_dropout.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier_dropout.add(Dropout(rate = 0.1))

# Adding the output layer
classifier_dropout.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# Compiling the ANN
classifier_dropout.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

**### Melatih Model ANN Baru**

```python
# Fitting the new ANN with Dropout to the Training set
classifier_dropout.fit(X_train, y_train, batch_size = 10, epochs = 20)
```

**### Membuat Prediksi dan Mengevaluasi Model ANN Baru**

```python
# Predicting the Test set results with the new model
y_pred_dropout = classifier_dropout.predict(X_test)
y_pred_dropout_ = (y_pred_dropout > 0.5)

# Making the Confusion Matrix for the new model
cm_dropout = confusion_matrix(y_test, y_pred_dropout_)
print("Confusion Matrix (with Dropout):\n", cm_dropout)

# Menghitung metrik performa untuk model baru
accuracy_dropout = accuracy_score(y_test, y_pred_dropout_)
precision_dropout = precision_score(y_test, y_pred_dropout_)
recall_dropout = recall_score(y_test, y_pred_dropout_)
f1_dropout = f1_score(y_test, y_pred_dropout_)

print("\n--- Performa Model ANN dengan Dropout ---")
print("Accuracy:", accuracy_dropout)
print("Precision:", precision_dropout)
print("Recall:", recall_dropout)
print("F1 Score:", f1_dropout)

print("\n--- Perbandingan dengan Model Asli ---")
print(f"Accuracy Asli: {accuracy:.4f}, Accuracy Dropout: {accuracy_dropout:.4f}")
print(f"Precision Asli: {precision:.4f}, Precision Dropout: {precision_dropout:.4f}")
print(f"Recall Asli: {recall:.4f}, Recall Dropout: {recall_dropout:.4f}")
print(f"F1 Score Asli: {f1:.4f}, F1 Score Dropout: {f1_dropout:.4f}")
```
