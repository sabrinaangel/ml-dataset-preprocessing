## Latihan: Variables

<b>Cerita Pendek:</b>

Suatu waktu di kebun apel, Darjo memiliki 3 apel, Darno memiliki 5 apel dan Waginem memiliki 6 Apel. mereka sangat bahagia dan hidup lama. selesai.

Tugas Anda
- buat variabe : `Darjo`, `Darno` dan `Waginem`
- isi nilai sesuai dengan jumlah apel yang mereka miliki
- cetak nama variabel dan jumlah apel setiap variabel dalam 1 baris, pisahkan dengan koma
- buat variabel baru dengan nama `totalApel` yang berisi penjumlahan seluruh apel yang mereka miliki
- coba otak atik code dengan membuat variable baru, diisi dengan nilai lain, dan dihitung dengan operator aritmetik lainnya

````python
Darjo = 2
Darno = 5
Waginem = 6
````

````python
# 1. Buat variabel dengan jumlah apel masing-masing
Darjo = 3
Darno = 5
Waginem = 6

# 2. Cetak nama variabel dan jumlah apel dalam 1 baris, pisahkan dengan koma
print("Darjo =", Darjo, ", Darno =", Darno, ", Waginem =", Waginem)

# 3. Buat variabel totalApel (penjumlahan seluruh apel)
totalApel = Darjo + Darno + Waginem
print("Total apel =", totalApel)

# 4. Otak-atik dengan variabel baru dan operator aritmetik lainnya
print("\n--- Otak-atik Code ---")

# Variabel baru
Mamat = 2  # Datang membawa 2 apel
Siti = 4   # Datang membawa 4 apel

# Penjumlahan
semuaApel = totalApel + Mamat + Siti
print("Total apel setelah Mamat dan Siti datang =", semuaApel)

# Pengurangan (apel dimakan)
apelDimakan = 3
sisaApel = semuaApel - apelDimakan
print(f"Setelah dimakan {apelDimakan} apel, sisa = {sisaApel}")

# Perkalian (apel dilipatgandakan)
apelAjaib = Darjo * 2  # Apel Darjo jadi 2x lipat
print("Apel Darjo setelah dilipatgandakan =", apelAjaib)

# Pembagian (apel dibagi rata)
rataRata = totalApel / 3  # Dibagi 3 orang
print("Rata-rata apel per orang =", rataRata)

# Modulo/sisa bagi
sisaBagi = totalApel % 2
print("Sisa bagi total apel dengan 2 =", sisaBagi)

# Pangkat
kuadrat = Darjo ** 2  # Apel Darjo pangkat 2
print("Apel Darjo pangkat 2 =", kuadrat)
````

## Latihan: Variables: a simple converter

Mil dan kilometer adalah satuan panjang

1 mil memiliki panjang sekitar 1.61 kilometer, buatlah program konversi di bawah ini
- mil ke kilometer;
- kilometer to mil.
jangan ganti apapun terhadap kode yang sudah ada. tulis kodemu pada tanda ###, kemudian hapus tanda tersebut. Uji kode Anda dengan data yang kami sajikan dalam source code

Kode:
```python
kilometer = 12.25
mil = 7.38

mil_ke_kilometer = ###
kilometer_ke_mil = ###

print(mil, "1 mil adalah", round(miles_to_kilometers, 2), "kilometer")
print(kilometer, "1 kilometer adalah", round(kilometers_to_miles, 2), "mil")
```

````python
# 1 mil = 1.61 kilometer
# Program konversi mil ke kilometer dan kilometer ke mil

# Konversi mil ke kilometer
mil = 10  # contoh: 10 mil
konversi_ke_km = mil * 1.61
print(f"{mil} mil = {konversi_ke_km} kilometer")

# Konversi kilometer ke mil
kilometer = 20  # contoh: 20 kilometer
konversi_ke_mil = kilometer / 1.61
print(f"{kilometer} kilometer = {konversi_ke_mil} mil")

# Atau dengan format yang lebih rapi
print("\n--- Program Konversi ---")

# Input dari user
print("Pilih konversi:")
print("1. Mil ke Kilometer")
print("2. Kilometer ke Mil")
pilihan = input("Masukkan pilihan (1/2): ")

if pilihan == "1":
    mil = float(input("Masukkan nilai mil: "))
    km = mil * 1.61
    print(f"{mil} mil = {km} kilometer")
elif pilihan == "2":
    km = float(input("Masukkan nilai kilometer: "))
    mil = km / 1.61
    print(f"{km} kilometer = {mil} mil")
else:
    print("Pilihan tidak valid")
````

````python
kilometer = 12.25
mil = 7.38

miles_to_kilometers = mil * 1.61
kilometers_to_miles = kilometer / 1.61

print(mil, "1 mil adalah", round(miles_to_kilometers, 2), "kilometer")
print(kilometer, "1 kilometer adalah", round(kilometers_to_miles, 2), "mil")
````
